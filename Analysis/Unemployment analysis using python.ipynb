```python
import pandas as pd
import plotly.express as px

unemployment = pd.read_csv("C:/Users/DELL/Downloads/Unemployment in India.csv")

unemployment.columns = unemployment.columns.str.strip()


qunemployment = unemployment.groupby(['Region', 'Frequency'])['Estimated Unemployment Rate (%)'].mean().reset_index()

figure = px.sunburst(
    qunemployment,
    path=['Region', 'Frequency'],
    values='Estimated Unemployment Rate (%)',
    width=700,
    height=600,
    color='Estimated Unemployment Rate (%)',
    color_continuous_scale="RdYlGn",
    title="Unemployment Rate in INDIA 2020"
)

figure.show()
```
